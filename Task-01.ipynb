```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exponential_array = np.random.exponential(scale=1, size=(4, 4))
```

```python
print(exponential_array)
```

```python
# Create a random 100000x1 array from exponential distribution
exponential_data = np.random.exponential(scale=1, size=100000)

# Create a random 100000x1 array from uniform distribution
uniform_data = np.random.rand(100000)

# Create a random 100000x1 array from normal distribution
normal_data = np.random.normal(size=100000)

# Plot histograms
plt.hist(exponential_data, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=50, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=50, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

# Show the plot
plt.show()
```
